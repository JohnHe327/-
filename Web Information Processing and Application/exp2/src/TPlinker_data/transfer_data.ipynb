```python
import re
import json

train_name = 'train.txt'
output_train_name = 'train_data.json'
valid_name = 'valid.txt'
output_valid_name = 'valid_data.json'
test_name = 'test.txt'
output_test_name = 'test_data.json'

def transfer_data(input_name, output_name, mode='train'):
    output_file = open(output_name, 'w')
    with open(input_name) as input_file:
        row_type = 'text'
        text = ""
        for row in input_file:
            if row_type == 'text':
                text = row.strip('1234567890 ')[:-1]
                row_type = 'relation'
            else:
                relation_end = row.index('(')
                relation = row[:relation_end]
                subject_end = row.index(',')
                subject = row[relation_end + 1:subject_end]
                object = row[subject_end + 1 : -2]
                
                data_dict = {}
                data_dict['sentText'] = text
                data_dict['relationMentions'] = []
                relation_dict = {}
                relation_dict['em1Text'] = subject
                relation_dict['em2Text'] = object
                relation_dict['label'] = relation
                data_dict['relationMentions'].append(relation_dict)

                json.dump(data_dict, output_file)
                output_file.write('\n')
                row_type = 'text'

    output_file.close()
```

```python
transfer_data(train_name, output_train_name, mode='train')
transfer_data(valid_name, output_valid_name, mode='valid')
transfer_data(test_name, output_test_name, mode='test')
```
